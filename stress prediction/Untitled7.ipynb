```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load the dataset directly
df = pd.read_csv('/content/lifestyle_dataset.csv')  # Change path if needed
```

```python
# Step 3: Display basic information
print("Dataset shape:", df.shape)
print("\nDataset columns:\n", df.columns)
print("\nFirst 5 rows:\n", df.head())
```

```python
# Step 4: Check for missing values
print("\nMissing values:\n", df.isnull().sum())
```

```python
# Step 5: Basic statistics
print("\nDescriptive statistics:\n", df.describe())

# Step 6: Count of Sleep Disorders (if column exists)
if 'Sleep Disorder' in df.columns:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x='Sleep Disorder', palette='Set2')
    plt.title("Count of Sleep Disorders")
    plt.show()
```

```python
# Step 7: Distribution of Sleep Duration by Sleep Disorder (if columns exist)
if 'Sleep Disorder' in df.columns and 'Sleep Duration' in df.columns:
    plt.figure(figsize=(8,5))
    sns.boxplot(data=df, x='Sleep Disorder', y='Sleep Duration', palette='Set3')
    plt.title("Sleep Duration by Sleep Disorder")
    plt.show()
```

```python
# Step 8: Correlation heatmap for numeric columns
plt.figure(figsize=(10,8))
numeric_cols = df.select_dtypes(include=['float64','int64']).columns
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
